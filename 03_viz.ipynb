# Python Visualizations: It all starts with Matplotlib otherwise it's Javascript

**Historical Context about Matplotlib:**

- Came to light in 2002
- Developped to mimic Matlab graphing, hence the name
- Matplotlib is the base layer for most python ploting libraries


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pivottablejs import pivot_ui
from ipywidgets import *
from ipywidgets import widget
from IPython.display import display;
```

```python
'''
load the car csv file into dataframe. 
'''
car_df=pd.read_csv("car_data.csv")
```

```python
pivot_ui(car_df)
```


**"The plot method on Series and DataFrame is just a simple wrapper around Matplotlib plt.plot()"**

For further readings on Pandas Visualization go to the following:

https://pandas.pydata.org/pandas-docs/stable/visualization.html

```python
car_df["speed"].plot();
```

```python
'''
plotting multiple columns, like speed and engine coolant
'''
car_df[["speed","coolant_temp"]].plot();
```

```python
'''
scatter Plotting Vehicle speed vs Engine RPM. 
'''
car_df.plot.scatter("speed","rpm");
```

**Matplotlib subplots allow to mix different types of plots**

```python
data=[car_df["speed"],car_df["rpm"]]
fig, axs = plt.subplots(2, 2, figsize=(7, 7))
axs[0, 0].hist(data[0])
axs[1, 0].scatter(data[0], data[1])
axs[0, 1].plot(data[0], data[1])
axs[1, 1].hist2d(data[0], data[1])
plt.show();
```

```python
from bokeh.plotting import figure, output_notebook, show, gridplot
from bokeh.layouts import row
from bokeh.models.glyphs import Circle, Line
from bokeh.models import ColumnDataSource
```

```python
output_notebook()
```

The command ``output_notebook()`` allows bokeh to output a graph/plot inside the Notebook

```python
'''
creating a new plot with a title and axis labels. 
'''
TOOLS="hover,crosshair,pan,wheel_zoom,box_zoom,reset,tap,save,box_select,poly_select,lasso_select"
p1 = figure(tools=TOOLS,plot_width=700, plot_height=700,title="Speed vs RPM", x_axis_label='Speed', 
            y_axis_label='RPM')
p1.scatter(car_df["speed"], car_df["rpm"])
show(p1);
```

```python
def plotwithwidget(speedlevel=10):
    df=car_df[car_df["speed"]>speedlevel]
    TOOLS="hover,crosshair,pan,wheel_zoom,box_zoom,reset,tap,save,box_select,poly_select,lasso_select"
    p1 = figure(tools=TOOLS,plot_width=500, plot_height=500,title="Speed vs RPM", x_axis_label='Speed', 
                y_axis_label='RPM')
    p1.scatter(df["speed"], df["rpm"])
    show(p1);
```

```python
widgets.interact(plotwithwidget,speedlevel=(20,60,5));
```
